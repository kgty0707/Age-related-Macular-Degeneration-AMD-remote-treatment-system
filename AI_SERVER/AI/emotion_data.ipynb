```python
import pandas as pd
import os

directory_path = './KEMDy20_v1_2/KEMDy20_v1_2/annotation'
total_emotions = pd.Series(dtype=int)
target_emotions = ['happy;sad']

for file in os.listdir(directory_path):
    if file.endswith('.csv'):
        file_path = os.path.join(directory_path, file)
        df = pd.read_csv(file_path)
        
        # Access the 4th column data (assuming k=4, zero-indexed, so it is the 5th column)
        k = 4
        column_data = df.iloc[:, k]  # Accessing the data in the 5th column
        print(f"Data from the {k+1}th column: {column_data.head()}")  # Print the first few rows
        
        # Filter for target emotions and count them
        filtered_data = column_data[column_data.isin(target_emotions)]
        if not filtered_data.empty:
            break
        emotion_counts = filtered_data.value_counts()
        total_emotions = total_emotions.add(emotion_counts, fill_value=0)

total_emotions_df = pd.DataFrame(total_emotions, columns=['Count'])
print(total_emotions_df)
```
